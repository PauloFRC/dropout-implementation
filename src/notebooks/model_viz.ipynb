## Visualization of Used Models

```python
from nn_models.simple import SimpleDropoutNetwork
from nn_models.lenet import DropoutLeNet
from nn_models.vgg import DropoutVGG
import torch

from torchview import draw_graph
```

```python
model = SimpleDropoutNetwork()

model_graph = draw_graph(
    model, 
    input_size=(1, 1, 28, 28),
    expand_nested=True,
    depth=3,
    device='cpu'
)

model_graph.visual_graph.render('simple_network', format='png', cleanup=True)
```
